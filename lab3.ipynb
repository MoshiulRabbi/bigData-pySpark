```python
from pyspark.sql.types import *
```

```python
def casesHighLow(confirmed):
    if confirmed < 50:
        return 'low'
    else:
        return 'High'



casesHighLowUdf = F.udf(casesHighLow, StringType())

casesWithHighLow = cases.withColumn('HighLow',casesHighLowUdf("confirmed"))

casesWithHighLow.show()
```

```python
timeprovince = spark.read.load("/home/eternal/bigData/Data/TimeProvince.csv", format = "csv",sep = ",", inferScheme = True, header = True )
```

```python
timeprovince.show()
```

```python
from pyspark.sql.window import Window
windowFrame = Window().partitionBy(["province"]).orderBy("Date")
timeProvincingwithLag = timeprovince.withColumn("lag_7",F.lag("confirmed",7).over(windowFrame))
timeProvincingwithLag.show()
```

```python
timeProvincingwithLag.filter(timeProvincingwithLag.date>'2020-03-10').show()
```
